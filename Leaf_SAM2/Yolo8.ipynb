```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
```

```python
model.train(
    data="/content/drive/MyDrive/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16
)
```

```python
model.val()
```

```python
results = model.predict(
    source="/content/drive/MyDrive/test/1.X5.1.888.jpg",
    conf=0.25,
    save=True
)
```

```python
model.export(format="onnx")
```
